```python
import nest_asyncio
nest_asyncio.apply()

import logging
from telegram import Update, ReplyKeyboardMarkup
from telegram.ext import (
    Application,
    CommandHandler,
    ContextTypes,
    MessageHandler,
    filters
)

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –ª–æ–≥–∏—Ä–æ–≤–∞–Ω–∏—è
logging.basicConfig(
    format="%(asctime)s - %(name)s - %(levelname)s - %(message)s",
    level=logging.INFO
)
logger = logging.getLogger(__name__)

TOKEN = "8131554002:AAFdUNndeZ_CQETeA5_xKonYaQ5v-d0RXqU"  # –ó–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ —Ä–µ–∞–ª—å–Ω—ã–π —Ç–æ–∫–µ–Ω!

# –¢–µ–∫—Å—Ç—ã –¥–ª—è –∫–æ–º–∞–Ω–¥
ABOUT_TEXT = """–ü—Ä–∏–≤–µ—Ç! –Ø –±–æ—Ç –ö–ª—É–±–∞ –ø–æ–ª–∏—Ç–∏—á–µ—Å–∫–æ–≥–æ –∫–æ–Ω—Å–∞–ª—Ç–∏–Ω–≥–∞ –ú–ì–ò–ú–û. 

–ö–ü–ö –ú–ì–ò–ú–û —ç—Ç–æ –Ω–µ –ø—Ä–æ—Å—Ç–æ —Å—Ç—É–¥–µ–Ω—á–µ—Å–∫–æ–µ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ. –ú—ã ‚Äî –∫–æ–º–∞–Ω–¥–∞ –±—É–¥—É—â–∏—Ö –ø–æ–ª–∏—Ç–∏—á–µ—Å–∫–∏—Ö —Ç–µ—Ö–Ω–æ–ª–æ–≥–æ–≤, –∞–Ω–∞–ª–∏—Ç–∏–∫–æ–≤, —Å—Ç—Ä–∞—Ç–µ–≥–æ–≤ –∏ —É–ø—Ä–∞–≤–ª–µ–Ω—Ü–µ–≤, –∫–æ—Ç–æ—Ä—ã–µ —É–∂–µ —Å–µ–≥–æ–¥–Ω—è —É—á–∞—Ç—Å—è –≤–ª–∏—è—Ç—å –Ω–∞ –ø–æ–≤–µ—Å—Ç–∫—É –∏ –ø–æ–±–µ–∂–¥–∞—Ç—å –≤ —Å–∞–º—ã—Ö —Å–ª–æ–∂–Ω—ã—Ö –∫–∞–º–ø–∞–Ω–∏—è—Ö.

–ó–¥–µ—Å—å –Ω–µ—Ç —Å–ª—É—á–∞–π–Ω—ã—Ö –ª—é–¥–µ–π ‚Äî —Ç–æ–ª—å–∫–æ —Ç–µ, –∫—Ç–æ —Ö–æ—á–µ—Ç –Ω–µ –ø—Ä–æ—Å—Ç–æ –∏–∑—É—á–∞—Ç—å –ø–æ–ª–∏—Ç–∏–∫—É, –∞ —Ç–≤–æ—Ä–∏—Ç—å –µ—ë. 

–ú—ã –Ω–µ –∂–¥—ë–º –¥–∏–ø–ª–æ–º–æ–≤, —á—Ç–æ–±—ã –Ω–∞—á–∞—Ç—å –¥–µ–π—Å—Ç–≤–æ–≤–∞—Ç—å: –º—ã –ø–∏—à–µ–º —Å—Ü–µ–Ω–∞—Ä–∏–∏ –≤—ã–±–æ—Ä–æ–≤, —Ä–∞–∑–±–∏—Ä–∞–µ–º –∫–µ–π—Å—ã –º–∏—Ä–æ–≤—ã—Ö –∫–∞–º–ø–∞–Ω–∏–π, —É—á–∏–º—Å—è —É–ø—Ä–∞–≤–ª—è—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–æ–Ω–Ω—ã–º–∏ –ø–æ—Ç–æ–∫–∞–º–∏ –∏ –ø—Ä–µ–≤—Ä–∞—â–∞—Ç—å –∫—Ä–∏–∑–∏—Å—ã –≤ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏.

–ü–æ—á–µ–º—É –º—ã ‚Äî –±–æ–ª—å—à–µ, —á–µ–º –∫–ª—É–±?

üîπ –ú—ã ‚Äî —Å–µ–º—å—è. –ó–¥–µ—Å—å —Ü–µ–Ω—è—Ç –Ω–µ —Ç–æ–ª—å–∫–æ –∑–Ω–∞–Ω–∏—è, –Ω–æ –∏ —Ö–∞—Ä–∞–∫—Ç–µ—Ä. –í–º–µ—Å—Ç–µ –º—ã –ø—Ä–æ—Ö–æ–¥–∏–º —á–µ—Ä–µ–∑ –∂–∞—Ä–∫–∏–µ –¥–∏—Å–∫—É—Å—Å–∏–∏, –º–æ–∑–≥–æ–≤—ã–µ —à—Ç—É—Ä–º—ã –¥–æ –ø–æ–∑–¥–Ω–µ–π –Ω–æ—á–∏ –∏ –ø–µ—Ä–≤—ã–µ –ø–æ–±–µ–¥—ã ‚Äî –∞ –ø–æ—Å–ª–µ –ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞–µ–º –¥—Ä—É–≥ –¥—Ä—É–≥–∞ –Ω–∞ –ø—É—Ç–∏ –∫ –Ω–æ–≤—ã–º –≤–µ—Ä—à–∏–Ω–∞–º.

üîπ –ú—ã ‚Äî –ª–∞–±–æ—Ä–∞—Ç–æ—Ä–∏—è –ø–æ–ª–∏—Ç–∏—á–µ—Å–∫–∏—Ö —Ç–µ—Ö–Ω–æ–ª–æ–≥–∏–π. –ù–∞—à–∏ –ø—Ä–æ–µ–∫—Ç—ã ‚Äî –Ω–µ —Ç–µ–æ—Ä–∏—è, –∞ —Ä–µ–∞–ª—å–Ω–∞—è –ø—Ä–∞–∫—Ç–∏–∫–∞: –æ—Ç –≤—ã–±–æ—Ä–æ–≤ –ø—Ä–µ–∑–∏–¥–µ–Ω—Ç–∞ –∫—É—Ä—Å–∞ –¥–æ —á–µ–º–ø–∏–æ–Ω–∞—Ç–æ–≤ –ø–æ –ø–æ–ª–∏—Ç–∏—á–µ—Å–∫–æ–º—É –∫–æ–Ω—Å–∞–ª—Ç–∏–Ω–≥—É.

üîπ –ú—ã ‚Äî –Ω–µ—Ç–≤–æ—Ä–∫–∏–Ω–≥, –∫–æ—Ç–æ—Ä—ã–π —Ä–∞–±–æ—Ç–∞–µ—Ç. –í—ã—Å—Ç—É–ø–ª–µ–Ω–∏—è —ç–∫—Å–ø–µ—Ä—Ç–æ–≤, –º–∞—Å—Ç–µ—Ä-–∫–ª–∞—Å—Å—ã –æ—Ç —Ç–æ–ø-–∫–æ–Ω—Å—É–ª—å—Ç–∞–Ω—Ç–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ —É–∂–µ –º–µ–Ω—è—é—Ç –ø–æ–ª–∏—Ç–∏–∫—É –≤ –†–æ—Å—Å–∏–∏ –∏ –∑–∞ —Ä—É–±–µ–∂–æ–º.

–ß—Ç–æ —É –Ω–∞—Å –µ—Å—Ç—å?

üéØ¬´–ú–∞—Å—Ç–µ—Ä—Å–∫–∞—è –ø–æ–ª–∏—Ç–∏—á–µ—Å–∫–æ–≥–æ –∫–æ–Ω—Å–∞–ª—Ç–∏–Ω–≥–∞¬ª ‚Äî –∑–¥–µ—Å—å —É—á–∞—Ç –Ω–µ –ø—Ä–æ—Å—Ç–æ –ø–∏—Å–∞—Ç—å –∞–Ω–∞–ª–∏—Ç–∏–∫—É, –∞ —Å–æ–∑–¥–∞–≤–∞—Ç—å –Ω–∞—Ä—Ä–∞—Ç–∏–≤—ã, –∫–æ—Ç–æ—Ä—ã–µ –º–µ–Ω—è—é—Ç –º–Ω–µ–Ω–∏–µ –ª—é–¥–µ–π.

üéØ –ö–µ–π—Å-—á–µ–º–ø–∏–æ–Ω–∞—Ç ¬´–ö–∞–Ω–¥–∏–¥–∞—Ç¬ª ‚Äî –Ω–∞—Å—Ç–æ—è—â–∞—è –±–∏—Ç–≤–∞ —Å—Ç—Ä–∞—Ç–µ–≥–∏–π, –≥–¥–µ –ø—Ä–æ–≤–µ—Ä—è–µ—à—å —Å–µ–±—è –Ω–∞ –ø—Ä–æ—á–Ω–æ—Å—Ç—å –∏ –Ω–∞—Ö–æ–¥–∏—à—å –µ–¥–∏–Ω–æ–º—ã—à–ª–µ–Ω–Ω–∏–∫–æ–≤.

–ö—Ç–æ —Ç—ã –≤ —ç—Ç–æ–º –º–∏—Ä–µ?

–ï—Å–ª–∏ —Ç—ã —á—É–≤—Å—Ç–≤—É–µ—à—å –∞–∑–∞—Ä—Ç, –∫–æ–≥–¥–∞ —Å–ª—ã—à–∏—à—å —Å–ª–æ–≤–æ ¬´–≤—ã–±–æ—Ä—ã¬ª, –µ—Å–ª–∏ —Ö–æ—á–µ—à—å –Ω–µ –ø—Ä–æ—Å—Ç–æ –Ω–∞–±–ª—é–¥–∞—Ç—å –∑–∞ –ø–æ–ª–∏—Ç–∏–∫–æ–π, –∞ –ª–µ–ø–∏—Ç—å –µ—ë —Å–≤–æ–∏–º–∏ —Ä—É–∫–∞–º–∏ ‚Äî –¥–æ–±—Ä–æ –ø–æ–∂–∞–ª–æ–≤–∞—Ç—å.–ú—ã –Ω–µ –≥–æ—Ç–æ–≤–∏–º –∫ –±—É–¥—É—â–µ–º—É. –ú—ã —Å–æ–∑–¥–∞—ë–º –µ–≥–æ —É–∂–µ —Å–µ–≥–æ–¥–Ω—è."

–ü—Ä–∏—Å–æ–µ–¥–∏–Ω—è–π—Å—è. –ò–≥—Ä–∞ —Ç–æ–ª—å–∫–æ –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è."""

JOIN_TEXT = """–ú—ã —Ä–∞–¥—ã, —á—Ç–æ —Ç—ã —Ö–æ—á–µ—à—å —Å—Ç–∞—Ç—å —á–∞—Å—Ç—å—é –Ω–∞—à–µ–π –∫–æ–º–∞–Ω–¥—ã. –ü–µ—Ä–µ—Ö–æ–¥–∏ –ø–æ —Å—Å—ã–ª–∫–µ - https://forms.yandex.ru/u/66f93dd7eb6146c196a36810/ –∏ –∑–∞–ø–æ–ª–Ω—è–π —Ñ–æ—Ä–º—É. –°–∫–æ—Ä–æ —Å —Ç–æ–±–æ–π —Å–≤—è–∂–µ—Ç—Å—è –Ω–∞—à –≥–ª–∞–≤–∞ HR –∏ –∞–∫–∞–¥–µ–º–∏—á–µ—Å–∫–æ–≥–æ —Å–æ—Ç—Ä—É–¥–Ω–∏—á–µ—Å—Ç–≤–∞!"""

BOOKS_TEXT = "–°—Å—ã–ª–∫–∞ –Ω–∞ –ø–æ–ª–µ–∑–Ω—É—é –ª–∏—Ç–µ—Ä–∞—Ç—É—Ä—É..."

PARTNERSHIP_TEXT = """–•–æ—Ç–∏—Ç–µ —Å—Ç–∞—Ç—å —á–∞—Å—Ç—å—é –Ω–∞—à–µ–≥–æ –¥–≤–∏–∂–µ–Ω–∏—è? –ú—ã –æ—Ç–∫—Ä—ã—Ç—ã –∫ –ø–∞—Ä—Ç–Ω—ë—Ä—Å—Ç–≤—É —Å –∫–æ–º–ø–∞–Ω–∏—è–º–∏, —ç–∫—Å–ø–µ—Ä—Ç–∞–º–∏ –∏ –æ—Ä–≥–∞–Ω–∏–∑–∞—Ü–∏—è–º–∏!

üîπ –î–ª—è –∫–æ–≥–æ?

1)–°—Ç—É–¥–µ–Ω—á–µ—Å–∫–∏–µ –∫–ª—É–±–∞ –∏–∑ —Ä–∞–∑–Ω—ã—Ö —É–Ω–∏–≤–µ—Ä—Å–∏—Ç–µ—Ç–æ–≤
2)–ë–∏–∑–Ω–µ—Å –∏ –ù–ö–û, –∑–∞–∏–Ω—Ç–µ—Ä–µ—Å–æ–≤–∞–Ω–Ω—ã–µ –≤ –ø–æ–ª–∏—Ç–∏—á–µ—Å–∫–æ–º –∞–Ω–∞–ª–∏–∑–µ –∏ GR
3)–ü–æ–ª–∏—Ç–∏—á–µ—Å–∫–∏–µ –∫–æ–Ω—Å—É–ª—å—Ç–∞–Ω—Ç—ã –∏ —Ç—Ä–µ–Ω–µ—Ä—ã, –≥–æ—Ç–æ–≤—ã–µ –¥–µ–ª–∏—Ç—å—Å—è –æ–ø—ã—Ç–æ–º
4)–ú–µ–¥–∏–∞ –∏ –∏–∑–¥–∞–Ω–∏—è, –∏—â—É—â–∏–µ –º–æ–ª–æ–¥—ã—Ö –∞–≤—Ç–æ—Ä–æ–≤ –∏ —ç–∫—Å–ø–µ—Ä—Ç–æ–≤
5)–ì–æ—Å—É–¥–∞—Ä—Å—Ç–≤–µ–Ω–Ω—ã–µ —É–ø—Ä–∞–≤–ª–µ–Ω—Ü—ã 

üîπ –ß—Ç–æ –ø—Ä–µ–¥–ª–∞–≥–∞–µ–º?

- –û—Ä–≥–∞–Ω–∏–∑–∞—Ü–∏—é —Å–æ–≤–º–µ—Å—Ç–Ω—ã—Ö –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏–π –∏ –∫–µ–π—Å-—á–µ–º–ø–∏–æ–Ω–∞—Ç–æ–≤
- –î–æ—Å—Ç—É–ø –∫ —Ç–∞–ª–∞–Ω—Ç–ª–∏–≤—ã–º –∞–Ω–∞–ª–∏—Ç–∏–∫–∞–º –∏ –±—É–¥—É—â–∏–º –ø—Ä–æ—Ñ–µ—Å—Å–∏–æ–Ω–∞–ª–∞–º –æ—Ç—Ä–∞—Å–ª–∏
- –£—á–∞—Å—Ç–∏–µ –≤ —ç–∫—Å–ø–µ—Ä—Ç–Ω—ã—Ö –¥–∏—Å–∫—É—Å—Å–∏—è—Ö –∏ –≤–æ—Ä–∫—à–æ–ø–∞—Ö
–î–∞–≤–∞–π—Ç–µ —Å–æ–∑–¥–∞–≤–∞—Ç—å —Å–∏–ª—å–Ω—ã–µ –ø—Ä–æ–µ–∫—Ç—ã –≤–º–µ—Å—Ç–µ! ‚ú®

–ù–∞–ø–∏—à–∏—Ç–µ –Ω–∞–º ‚Äî –æ–±—Å—É–¥–∏–º –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏! –°–≤—è–∑–∞—Ç—å—Å—è –ø–æ –ø–æ–≤–æ–¥—É —Å–æ—Ç—Ä—É–¥–Ω–∏—á–µ—Å—Ç–≤–∞ –º–æ–∂–Ω–æ —á–µ—Ä–µ–∑ —Å—Å—ã–ª–∫—É"""

SOCIAL_TEXT = "–ù–∞—à–∏ —Å–æ—Ü—Å–µ—Ç–∏:..."

async def start(update: Update, context: ContextTypes.DEFAULT_TYPE):
    # –£–ø—Ä–æ—â–µ–Ω–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è –∫–Ω–æ–ø–æ–∫ (–±–µ–∑ –∫–æ–º–∞–Ω–¥ –≤ —Å–∫–æ–±–∫–∞—Ö)
    keyboard = [
        ["–û –∫–ª—É–±–µ", "–í—Å—Ç—É–ø–∏—Ç—å"],
        ["–ö–Ω–∏–≥–∏", "–°–æ—Ç—Ä—É–¥–Ω–∏—á–µ—Å—Ç–≤–æ"],
        ["–°–æ—Ü—Å–µ—Ç–∏"]
    ]
    reply_markup = ReplyKeyboardMarkup(keyboard, resize_keyboard=True, one_time_keyboard=False)
    await update.message.reply_text(
        "–î–æ–±—Ä–æ –ø–æ–∂–∞–ª–æ–≤–∞—Ç—å! –í—ã–±–µ—Ä–∏—Ç–µ –∏–Ω—Ç–µ—Ä–µ—Å—É—é—â–∏–π –≤–∞—Å —Ä–∞–∑–¥–µ–ª:",
        reply_markup=reply_markup
    )

async def about(update: Update, context: ContextTypes.DEFAULT_TYPE):
    await update.message.reply_text(ABOUT_TEXT)

async def join(update: Update, context: ContextTypes.DEFAULT_TYPE):
    await update.message.reply_text(JOIN_TEXT)

async def books(update: Update, context: ContextTypes.DEFAULT_TYPE):
    await update.message.reply_text(BOOKS_TEXT)

async def partnership(update: Update, context: ContextTypes.DEFAULT_TYPE):
    await update.message.reply_text(PARTNERSHIP_TEXT)

async def social(update: Update, context: ContextTypes.DEFAULT_TYPE):
    await update.message.reply_text(SOCIAL_TEXT)

def run_bot():
    app = Application.builder().token(TOKEN).build()
    
    # –†–µ–≥–∏—Å—Ç—Ä–∏—Ä—É–µ–º –æ–±—Ä–∞–±–æ—Ç—á–∏–∫–∏ –∫–æ–º–∞–Ω–¥
    app.add_handler(CommandHandler("start", start))
    app.add_handler(CommandHandler("about", about))
    app.add_handler(CommandHandler("join", join))
    app.add_handler(CommandHandler("books", books))
    app.add_handler(CommandHandler("partnership", partnership))
    app.add_handler(CommandHandler("social", social))
    
    # –û–±—Ä–∞–±–æ—Ç—á–∏–∫–∏ –¥–ª—è —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö —Å–æ–æ–±—â–µ–Ω–∏–π (—Ç–æ—á–Ω–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç —Ç–µ–∫—Å—Ç—É –∫–Ω–æ–ø–æ–∫)
    app.add_handler(MessageHandler(filters.TEXT & filters.Regex(r'^–û –∫–ª—É–±–µ$'), about))
    app.add_handler(MessageHandler(filters.TEXT & filters.Regex(r'^–í—Å—Ç—É–ø–∏—Ç—å$'), join))
    app.add_handler(MessageHandler(filters.TEXT & filters.Regex(r'^–ö–Ω–∏–≥–∏$'), books))
    app.add_handler(MessageHandler(filters.TEXT & filters.Regex(r'^–°–æ—Ç—Ä—É–¥–Ω–∏—á–µ—Å—Ç–≤–æ$'), partnership))
    app.add_handler(MessageHandler(filters.TEXT & filters.Regex(r'^–°–æ—Ü—Å–µ—Ç–∏$'), social))
    
    # –û–±—Ä–∞–±–æ—Ç—á–∏–∫ –¥–ª—è –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω—ã—Ö –∫–æ–º–∞–Ω–¥
    app.add_handler(MessageHandler(filters.TEXT, lambda update, ctx: update.message.reply_text("–ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫–Ω–æ–ø–∫–∏ –º–µ–Ω—é –∏–ª–∏ –∫–æ–º–∞–Ω–¥—ã.")))
    
    app.run_polling()

# –ó–∞–ø—É—Å–∫ –±–æ—Ç–∞
try:
    run_bot()
except KeyboardInterrupt:
    print("–ë–æ—Ç –æ—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω")
except Exception as e:
    logger.error(f"–û—à–∏–±–∫–∞ –≤ —Ä–∞–±–æ—Ç–µ –±–æ—Ç–∞: {e}")
```
